# An√°lise de Dados com pandas e Scikit-Learn

```python
# Conerter dados do MongoDB para o formato do Pandas (DataFrame)
# Data Frame √© como um excell, √© um arquivo de dados no formato de linhas e colunas
dataset = [{"criado_em":item["criado_em"], "texto":item["texto"],} for item in col.find()]
```

```python
# Importar pandas para trabalhar com dataset em python
import pandas as pd
```

```python
# Para o pandas trabalhar deve ser convertido o objeto para o tipo DataFrame
df = pd.DataFrame(dataset)
```

```python
df
```

```python
# Scikit Learn √© o melhor modulo para trabalhar com Machine Learning
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# CountVectorizer √© utilizado para contar quantas ocorrencias de uma palavera existem em um dataset
# O CountVectorizer cria uma matriz de dados, que contabiliza quantas vezes cada palavra aparece
# Deve-se instanciar a classe
cv = CountVectorizer()
count_matrix = cv.fit_transform(df.texto)
```

```python
# Contar o numero de ocorrencias das principais palavras no dataset
# conta_palavras recebe um DataFrame, que est√° recebendo a funcao get_feature_names do contador de palavras do scikit Learn, definimos o titulo da coluna como "palavra"
conta_palavra = pd.DataFrame(cv.get_feature_names(), columns=["palavra"])

# Definimos uma coluna "contador" que recebe a soma da matrix, sendo que deve ser convertida para lista com o tolist()
conta_palavra["contador"] = count_matrix.sum(axis=0).tolist()[0]

# Ordenamos os valores com sort_values
conta_palavra = conta_palavra.sort_values("contador", ascending=False).reset_index(drop=True)

# Retorna as primeiras 50 palavras
conta_palavra[:50]
```
