## Pylab

```python
# O Pylab combina funcionalidades do NumPy e do Pyplot
# Visualizando Graficos dentro do Jupyter
import pylab as plb
%matplotlib inline
```

```python
x = plb.linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

# Definindo os Eixos
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8 ])

axes.plot(x, y, 'r')


axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de Linha');
```

```python
# Grafico com 2 figuras
x = plb.linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8 ]) # Eixos da fugura principal
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3 ]) # Eixos da figura secundaria

# Figura Principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Fig Principal')


# Figura Segundaria
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Fig Secundaria')
```

```python
# Graficos em Paralelo
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Titulo')
    
fig.tight_layout()
```

## Graficos a partir do NumPy

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
valores1 = np.arange(50)
valores2 = np.random.randn(50)
plt.scatter(valores1, valores2)
plt.show()
```

```python
# Plot e Scatter
fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)
valores1 = np.random.randn(50)
ax1.plot(valores1, color='red')

ax2 = fig.add_subplot(1,2,2)
valores2 = np.arange(50)
valores3 = np.random.randn(50)
ax2.scatter(valores2, valores3)

plt.show()
```

```python
# Plots Diversos
_, ax = plt.subplots(2, 3)

valores1 = np.random.randn(50)
valores2 = np.arange(50)

ax[0, 1].plot(valores1, color='green', linestyle='-')
ax[1, 0].hist(valores1)
ax[1, 2].scatter(valores2, valores1, color='red')

plt.show()
```

```python
# Controle dos Eixos
fig, axes = plt.subplots(1, 3, figsize = (12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title('Eixos com Range Padrao')

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title('Eixos Menores')

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title('Eixos Customizados')
```

```python
# Escala
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title('Escala Padrao')

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale('log')
axes[1].set_title('Escala Logaritmica (y)')
```

```python
# Grid
fig, axes = plt.subplots(1, 2, figsize=(10, 3))

# Grid Padrao
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

# Grid Customizado
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
```

```python
# Grafico Linhas Gemeas
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color='blue')
ax1.set_ylabel('Area', fontsize=18, color='blue' )
for label in ax1.get_yticklabels():
    label.set_color('blue')
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color='red')
ax1.set_ylabel('Volume', fontsize=18, color='red')
for label in ax2.get_yticklabels():
    label.set_color('red')
```

```python
# Diferentes estilos de Plot
xx = np.linspace(-0.75, 1., 100)
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(12, 3))

axes[0].scatter(xx, xx + 0.25 * plb.randn(len(xx)))
axes[0].set_title('Scatter')

axes[1].step(n, n**2, lw=2)
axes[1].set_title('Step')

axes[2].bar(n, n**2, align='center', width=0.5, alpha=0.5)
axes[2].set_title('Bar')

axes[3].fill_between(x, x**2, x**3, color='green', alpha=0.5)
axes[3].set_title('Fill Between')
```

```python
# Histogramas
n = np.random.randn(100000)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(n)
axes[0].set_title('Histograma Padrao')
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title('Histograma Cuumulativo')
axes[1].set_xlim((min(n), max(n)))
```
