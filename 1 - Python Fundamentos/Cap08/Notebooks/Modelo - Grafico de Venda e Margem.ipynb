```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
periodo = pd.date_range('01/03/2019', periods=5, freq='D')
valorVenda = [round(x, 2) for x in np.random.random_sample(5) * 9.65]
margem = np.random.randint(15, 25, len(valorVenda))
```

```python
# Junta os 2 objetos
x = np.stack((valorVenda, margem))
```

```python
labels = ['Valor Venda', 'Margem']
```

## Stack 

```python
fig, ax = plt.subplots()
ax.stackplot(periodo, valorVenda, margem, labels=labels)
ax.legend(loc='upper left')
plt.show()
```

## Barchart

```python
import numpy as np
import matplotlib.pyplot as plt


n_groups = 5

means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)

means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = ax.bar(index, means_men, bar_width,
                alpha=opacity, color='b',
                yerr=std_men, error_kw=error_config,
                label='Men')

rects2 = ax.bar(index + bar_width, means_women, bar_width,
                alpha=opacity, color='r',
                yerr=std_women, error_kw=error_config,
                label='Women')

ax.set_xlabel('Group')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(('A', 'B', 'C', 'D', 'E'))
ax.legend()

fig.tight_layout()
plt.show()
```

```python
periodo = pd.date_range('01/03/2019', periods=5, freq='D')
valorVenda = [round(x, 2) for x in np.random.random_sample(5) * 9.65]
margem = np.random.randint(15, 25, len(valorVenda))
```

```python
fig, ax = plt.subplots()
index = np.arange(len(periodo))
bar_width = 0.35
opacity=0.4

rects1 = ax.bar(index, valorVenda, bar_width,
               alpha=opacity, color='b',
               label='Valor Venda')

rects2 = ax.bar(index, margem, bar_width,
               alpha=opacity, color='r',
               label='Margem')

ax.set_xlabel('Periodo')
ax.set_ylabel('Valores')
ax.set_title('Venda X Margem')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(periodo, rotation=75)
ax.legend()


fig.tight_layout()
plt.show()
```
